```python
data = {"x": 10, "y": 20, "z": 30, "t": 15}
```

```python
class Data:

    def __init__(self, x, y, z, t):
        self.x = x
        self.y = y
        self.z = z
        self.t = t

    # def __str__(self):
    #    return f"x: {self.x}, y: {self.y}, z: {self.z}, t: {self.t}"

    def calc(self):
        return self.x + self.y + self.z + self.t
```

```python
from dataclasses import dataclass

@dataclass
class Data1:
    x: int
    y: int
    z: int
    t: int

    def calc(self):
        return self.x + self.y + self.z + self.t
```

```python
d = Data(x=10, y=20, z=30, t=15)
d1 = Data1(x=10, y=20, z=31, t=16)
```
